```python
import Bio
```

```python
# PDB
from Bio.PDB.PDBParser import PDBParser

parser = PDBParser()
```

```python
# READ PDB File
structure_PDB = parser.get_structure('6lu7', file='6lu7.pdb')
print (structure_PDB)
```

```python
# Parse Information 
model_PDB = structure_PDB[0]
print (model_PDB)
```

```python
# Chain
for chain in model_PDB:
    print(f'Chain {chain}, ID{chain.id}')
```

```python
# Residu
for chain in model_PDB:
    print(f'Chain {chain}, ID{chain.id}')
    for residu in chain:
        print (residu)
```

```python
# Atom
for chain in model_PDB:
    print(f'Chain {chain}, ID{chain.id}')
    for residu in chain:
        print (residu)
        for atom in residu:
            print (atom)
```

```python
# Read File FASTA
from Bio import SeqIO
```

```python
# Method Parse
# Read File
fasta_parse = SeqIO.parse('coronavirus.fasta', 'fasta')

# Parse
for record in fasta_parse:
    print(record)
    print ("_______")
    print (record.id)
    print (record.description)
    print (record.features)
    print (record.seq)
```

```python
# Method Read
fasta_read = SeqIO.read('coronavirus.fasta', 'fasta')

# Parse
print (fasta_read)
```

```python
# GenBank
genbank_parse = SeqIO.parse('coronavirus.gb', 'gb')

# Parse
for record in genbank_parse:
    print (record)
    print ("______________")
    print (record.features)
    print (record.seq)
```

```python
# Read
genbank_read = SeqIO.read('coronavirus.gb', 'gb')

# Parse
print (genbank_read)
```

```python
print (genbank_read.seq.translate())

split_seq = genbank_read.seq.translate().split('*')

for protein in split_seq:
    print(f'Protein : {protein}, Lenght {len(protein)}')
```

```python
# Split Sequence
Seq_Source = genbank_read.seq

# SeqA -> 0 - 1/3
SeqA = Seq_Source[: round (len(Seq_Source)/3)]
# SeqB -> 1/3 - 2/3
SeqB = Seq_Source[ round (len(Seq_Source)/3) : round(len(Seq_Source)*2/3)]

print (len(SeqA))
```

```python
# Import Melting Temperature
from Bio.SeqUtils import MeltingTemp as mt
```

```python
# Calculate MT for each Sequence
mt_SeqA = mt.Tm_GC(SeqA)
print ("Temp A = ", mt_SeqA)

mt_SeqB = mt.Tm_GC(SeqB)
print ("Temp B = ", mt_SeqB)
```

```python
import matplotlib.pyplot as plt
```

```python
# Plotting
plt.subplot(1, 2, 1)
plt.bar(['SeqA', 'SeqB'], [mt_SeqA, mt_SeqB])
plt.show()
```
